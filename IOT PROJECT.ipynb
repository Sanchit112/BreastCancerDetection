### Training-

```python
import pandas as pd
import matplotlib as mp
import numpy as np
import seaborn as sb
%matplotlib inline
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.grid_search import GridSearchCV
```

```python
cancer= load_breast_cancer()
```

```python
df_features=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df=df_features
df['binary']=cancer['target']
df1=df[['mean radius',
 'mean texture',
 'mean perimeter',
 'mean area',
 'mean smoothness','binary']]

sb.pairplot(df1,hue='binary',palette="OrRd")
```

```python
f=df_features
d= cancer['target']
f_train, f_test, d_train, d_test = train_test_split(f, d, test_size=0.3, random_state=101)
```

```python
gamma_range = [10,1,0.1,0.01,0.001,0.0001]
C_range = [0.01,0.1,1,10,100,1000,10000]
param_grid = dict(gamma=gamma_range, C=C_range)
grid = GridSearchCV(SVC(), param_grid=param_grid,verbose=3)
grid.fit(f_train, d_train)
```

```python
grid.best_params_
```

__The Best Value for C is <font color=red>10</font><br>
The Best Value for gamma is <font color=red>0.0001</font>__

***
### Testing-

```python
grid_predictions=grid.predict(f_test)
```

```python
print(classification_report(d_test,grid_predictions))
```

```python
print(confusion_matrix(d_test,grid_predictions))
```
